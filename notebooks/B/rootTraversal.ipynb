# Using uproot to explore root files

```python
import uproot as ur
import awkward as ak
import numpy as np
```

First we gotta read the root file. If you already have info about what tree the info you wanna look at, you can include that and it will only read that tree.
The upshot here is that uproot will let you read root files as if they were python dictionaries. This object that is like a python dictionary is called an awkward array. What makes it more special than a regular dictionary is that you can do operations on it as if it were something like a numpy array.

```python
filename = 'root://cms-xrd-global.cern.ch//store/mc/Phase2HLTTDRSummer20ReRECOMiniAOD/TT_TuneCP5_14TeV-powheg-pythia8/FEVT/PU200_111X_mcRun4_realistic_T15_v1-v2/280000/003ACFBC-23B2-EA45-9A12-BECFF07760FC.root'
root = ur.open(filename)
```

We can list the keys to the dictionary

```python
root.keys()
```

We probably want to look at the subranches of 'Events'

```python
root['Events'].keys()
```

That is a lot of sub branches. Let's filter them.
